```python
import tensorflow as tf
from keras.models import Model
from keras.layers import Input, Dense

# Если X_train — scipy sparse matrix, преобразуем в tf.SparseTensor
def scipy_sparse_to_tf_sparse(scipy_sparse):
    coo = scipy_sparse.tocoo()
    indices = tf.constant(list(zip(coo.row, coo.col)), dtype=tf.int64)
    values = tf.constant(coo.data, dtype=tf.float32)
    shape = tf.constant(coo.shape, dtype=tf.int64)
    sparse_tensor = tf.sparse.SparseTensor(indices, values, shape)
    # Упорядочиваем индексы
    sparse_tensor = tf.sparse.reorder(sparse_tensor)
    return sparse_tensor

X_train_sparse = scipy_sparse_to_tf_sparse(X_train)

# Создаем модель с входом SparseTensor
input_layer = Input(shape=(X_train.shape[1],), sparse=True)
hidden = Dense(64, activation='relu')(input_layer)
output_layer = Dense(y_train.shape[1], activation='sigmoid')(hidden)

neur_model = Model(inputs=input_layer, outputs=output_layer)
neur_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Обучение модели
neur_model.fit(X_train_sparse, y_train, epochs=10, batch_size=32)
```
